```python
import pandas as pd

df = pd.read_csv("Sample - Superstore.csv", encoding="latin1")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df.dtypes
```

```python
import matplotlib.pyplot as plt

daily_sales = df.groupby('Order Date')['Sales'].sum()

plt.figure(figsize=(12,6))
plt.plot(daily_sales.index, daily_sales.values, linewidth=2)

plt.title("Sales Over Time")
plt.xlabel("Order Date")
plt.ylabel("Sales")
plt.legend(["Sales"])
plt.grid(True)

plt.savefig("sales_over_time.png", dpi=300)

plt.show()
```

```python
monthly_sales = df.resample('ME', on='Order Date')['Sales'].sum()

monthly_sales.head()
```

```python
plt.figure(figsize=(12,6))

plt.plot(monthly_sales.index,
         monthly_sales.values,
         marker='o',
         linewidth=2)

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend(["Monthly Sales"])
plt.grid(True)

plt.savefig("monthly_sales.png", dpi=300)

plt.show()
```

```python
quarterly_sales = df.resample('QE', on='Order Date')['Sales'].sum()

quarterly_sales
```

```python
plt.figure(figsize=(10,5))

plt.plot(quarterly_sales.index,
         quarterly_sales.values,
         marker='o',
         linewidth=2)

plt.title("Quarterly Sales Trend")
plt.xlabel("Quarter")
plt.ylabel("Sales")
plt.legend(["Quarterly Sales"])
plt.grid(True)

plt.savefig("quarterly_sales.png", dpi=300)

plt.show()
```

```python
category_sales = df.groupby('Category')['Sales'].sum()

category_sales
```

```python
plt.figure(figsize=(8,5))

category_sales.plot(kind='bar')

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)
plt.grid(axis='y')

plt.savefig("category_bar.png", dpi=300)

plt.show()
```

```python
plt.figure(figsize=(7,7))

category_sales.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90
)

plt.ylabel("")
plt.title("Sales Share by Category")

plt.savefig("category_pie.png", dpi=300)

plt.show()
```

```python
summary = pd.DataFrame({
    "Total Sales": [df["Sales"].sum()],
    "Average Sales": [df["Sales"].mean()],
    "Maximum Sale": [df["Sales"].max()],
    "Minimum Sale": [df["Sales"].min()],
    "Total Orders": [len(df)]
})

summary
```

```python
summary.to_csv("sales_summary.csv", index=False)
```
